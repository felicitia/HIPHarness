```python
import csv
import sys
import time as sys_time

csv.field_size_limit(sys.maxsize)
data_dir = '/Users/felicitia/Desktop/XXX/'

srcIP_idx = 0
desIP_idx = 1
host_idx = 2
url_idx = 3
time_idx = 4

ip_dict = dict()
host_dict = dict()

for i in range(7):
    start_time = sys_time.time()
    file_name = data_dir + str(i) + '.csv'
    print('processing ', file_name, '...')
    with open(file_name, encoding='utf-8') as csv_file:
        csv_reader = csv.reader(csv_file, delimiter='\x01')
        line_count = 0

        for row in csv_reader:
            ip = row[srcIP_idx]
            host = row[host_idx]
            
            if ip in ip_dict:
                ip_dict[ip] += 1
            else:
                ip_dict[ip] = 1
            if host in host_dict:
                host_dict[host] += 1
            else:
                host_dict[host] = 1
                
            line_count += 1
        print(f'Processed {line_count} lines.')
    print("--- %s seconds ---" % (sys_time.time() - start_time))

print('ip dict len = ', len(ip_dict))
print('host dict len = ', len(host_dict))
```

```python
# group top_ip and top_host entries

import csv
import sys
import time as sys_time

csv.field_size_limit(sys.maxsize)
data_dir = '/Users/felicitia/Desktop/XXX/'

srcIP_idx = 0
desIP_idx = 1
host_idx = 2
url_idx = 3
time_idx = 4

top_ip = list()
top_host = list()

sorted_ip = sorted(ip_dict.items(), key=lambda x: x[1], reverse=True)
print('top sorted ip', sorted_ip[0:10])
for tup in sorted_ip[0:10]:
    top_ip.append(tup[0])
print('top ip', top_ip)
print()
sorted_host = sorted(host_dict.items(), key=lambda x: x[1], reverse=True)
print('top host', sorted_host[0:10])
for tup in sorted_host[0:10]:
    top_host.append(tup[0])
print('top host', top_host)
print()

def writerow2csv(output_file, row):
    with open(output_file, "a") as output:
        writer = csv.writer(output, lineterminator='\n')
        writer.writerow(row)    
        
for i in range(7):
    start_time = sys_time.time()
    file_name = data_dir + str(i) + '.csv'
    print('processing ', file_name, '...')
    with open(file_name, encoding='utf-8') as csv_file:
        csv_reader = csv.reader(csv_file, delimiter='\x01')
        line_count = 0

        for row in csv_reader:
            ip = row[srcIP_idx]
            host = row[host_idx]
            if ip in top_ip:
                writerow2csv(data_dir + 'TopIP/' + ip + '.csv', row)
            if host in top_host:
                writerow2csv(data_dir + 'TopHost/' + host + '.csv', row)
                
            line_count += 1
        print(f'Processed {line_count} lines.')
    print("--- %s seconds ---" % (sys_time.time() - start_time))
```

```python
# sort files by timestamp

import csv
import glob
import sys
import time as sys_time

data_dir = '/Users/felicitia/Desktop/XXX/'
time_idx = 4

csv.field_size_limit(sys.maxsize)

def sortfile(input_file, output_file, idx):    
    with open(input_file) as csv_file:
        csv_reader = csv.reader(csv_file, delimiter=',')
        sortedlist = sorted(csv_reader, key=lambda row: row[idx])

    with open(output_file, "w") as output:
        writer = csv.writer(output, lineterminator='\n')
        for val in sortedlist:
            writer.writerow(val)  
            
print('processing top ip files...')
start_time = sys_time.time()
for file in glob.glob(data_dir + 'TopIP/*'):
    sortfile(file, file.replace('.csv', '') + '_sorted.csv', time_idx)
print("--- %s seconds ---" % (sys_time.time() - start_time))

print('processing top host files...')
start_time = sys_time.time()
for file in glob.glob(data_dir + 'TopHost/*'):
    sortfile(file, file.replace('.csv', '') + '_sorted.csv', time_idx)
print("--- %s seconds ---" % (sys_time.time() - start_time))    
```
